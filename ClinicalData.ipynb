```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("diabetes_dataset_with_notes.csv")
df
```

```python
plt.figure(figsize=(8,8))
sns.histplot(data=df, x="smoking_history", hue="gender" )
```

```python
df.drop(["clinical_notes"], axis=1, inplace=True)
df.describe()
```

```python
df.drop(["location"], axis=1, inplace=True)
df.info()
```

```python
df=pd.get_dummies(df["gender"]).join(df).drop(["gender"], axis=1)
df.info()
```

```python
df["Male"]= df["Male"].astype(int)
df["Other"]= df["Other"].astype(int)
df.head()
```

```python
df=pd.get_dummies(df["smoking_history"]).join(df).drop(["smoking_history"], axis=1)
df.head()
```

```python
df["never"]=df["never"].astype(int)
df["not current"]=df["not current"].astype(int)
df["current"]=df["current"].astype(int)
df["No Info"]=df["No Info"].astype(int)
df["ever"]=df["ever"].astype(int)
df["former"]=df["former"].astype(int)
df.head()
```

```python
features=df.drop(["diabetes", "year"], axis=1)
target= df["diabetes"]
features.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model=DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
results=pd.DataFrame({"actual":y_test, "predictions":y_pred})
results.reset_index(drop=True, inplace=True)
results.head()
```

```python
sns.histplot(data=results)
```
